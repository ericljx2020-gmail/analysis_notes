```python
import pandas as pd
import csv
```

```python
df = pd.read_parquet('/expanse/lustre/projects/sds193/magao/project_data/bgp_echo/rrc01/joined_route_dumps/d-1/2023-06/20230615.parquet.gzip')
```

```python
df['t_minus_3'] = df['date'] - pd.Timedelta(days=1)
df['t_plus_3'] = df['date'] + pd.Timedelta(days=1)
df.head()
```

```python
df_sub = df[['t_minus_3', 't_plus_3', 'victim_24']]
df_sub
```

```python
# Define the file path where you want to save the CSV file
csv_file_path = "./df_sub.csv"

# Save the DataFrame as a CSV file
df_sub.to_csv(csv_file_path, index=False)

print(f"DataFrame saved to {csv_file_path}")
```
